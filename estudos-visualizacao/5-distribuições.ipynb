```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("IRIS.csv")
df.head()
```

#### Histograma

- Visualizar a distribuição de frequência de uma variável numérica contínua

- Visualizar a distribuição dos dados ao longo de classes

- Visualização de características como simetria, assimetria, picos e outliers

- Comparação do dataset com uma distribuição teórica (Normal)

- Visualização da densidade dos dados

```python
sns.histplot(df["petal_length"])
```

#### Gráfico de Densidade

- Visualizar a distribuição contínua dos dados de forma suave, diferente do histograma (barras e intervalos discretos)

- Visualizar onde os dados estão mais concentrados

- Visualização da forma da distribuição

- Comparação de distribuições diferentes no mesmo gráfico (cores diferentes)

```python
# Eixo y em Histogramas = frequência absoluta ou relativa

# Eixo y em Gráfico de Densidade = frequência de probabilidade estimada em cada ponto contínuo do eixo x
# a área sob a toda a curva é igual a 1 (100% de probabilidade)
# a altura da curva mostra a concentração relativa dos dados, mas não é uma contagem direta 

sns.kdeplot(data=df["petal_length"], fill = True)
```

```python
# gráfico bidimensional
# 'kde' é 'kernel density estimate'
sns.jointplot(x=df["petal_length"], y = df["sepal_width"], kind="kde")

# o desenho em cima é o gráfico de "petal_length", o do lado direito é de "petal_width"
```

```python
# "hue" vai determinar as cores de cada grupo (cada espécie de pétala)
sns.histplot(data=df, x = "petal_length", hue="species")
plt.title("Histograma do Tamanho de Pétalas, por Espécie")
```

```python
# 'fill' é para preencher os espaços abaixo das curvas
sns.kdeplot(data=df, x="petal_length", hue="species", fill=True)
plt.title("Distribuição do Tamanho de Pétalas, por Espécie")
```
