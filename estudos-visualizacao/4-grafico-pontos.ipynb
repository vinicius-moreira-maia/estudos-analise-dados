```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv")
df.head()
```

#### Gráfico de pontos

- Visualização da relação entre 2 variáveis

- Visualização de padrões, tendências ou agrupamentos (clusters)

- Visualização de outliers

- Visualização da dispersão dos dados e a variabilidade dos mesmos

```python
sns.scatterplot(x=df['bmi'], y=df['charges'])

# correlação positiva é quando as variáveis tendem a ser diretamente proporcionais
```

```python
# adicionando uma Linha de Regressão
# a linha que "melhor se encaixa" nos dados
# essa linha vai me indicar se a correlação é positiva, negativa e o grau da correlação
sns.regplot(x=df['bmi'], y=df['charges'])
```

```python
# também posso mostrar a relação entre 3 variáveis, usando pontos coloridos
# 'hue' aqui usa uma coluna booleana que dá cor ao ponto dependendo do valor booleano
sns.scatterplot(x=df['bmi'], y=df['charges'], hue=df['smoker'])
```

```python
# adicionando 2 linhas de regressão
# 'hue' faz com que o seaborn crie uma linha para cada categoria
sns.lmplot(x='bmi', y='charges', hue='smoker', data=df)

# esse gráfico me mostra a linha de regressão para fumantes e não fumantes, a partir da relação entre IMC (bmi) e custos do plano de saúde
# para não-fumantes a correlação é quase que linear
# para fumantes os custos crescem bastante
```

```python
# gráfico de pontos também podem conter dados categóricos em algum dos eixos (não numéricos)
# isso me dá uma visão de distribuição das categorias ...
sns.swarmplot(x=df['smoker'], y=df['charges'])
```
