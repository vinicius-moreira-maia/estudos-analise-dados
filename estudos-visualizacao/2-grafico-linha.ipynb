```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# matplot > pyplot > seaborn
```

```python
# parsed_dates faz com que o Pandas "entenda" que cada linha se refere a uma data, não a um número de linha
df = pd.read_csv("WorldCups.csv", index_col=0, parse_dates=True)
df = df.iloc[:, [1, 5, 7, 8]]
#df.head()
df.tail()
```

```python
df.columns
```

```python
# dimensão da figura
plt.figure(figsize=(14, 6))

# título
plt.title("Números das Copas do Mundo")

#sns.lineplot(data=df)

# selecionando apenas 1 coluna para plotar, no lugar de todas
sns.lineplot(data=df["GoalsScored"], label="GoalsScored")

# esse comando desenha tudo que foi feito até esse ponto
# plt.show()

# selecionando apenas 1 coluna para plotar, no lugar de todas
sns.lineplot(data=df["MatchesPlayed"], label="MatchesPlayed")

# nome do eixo x
plt.xlabel("Ano da Copa")
```
