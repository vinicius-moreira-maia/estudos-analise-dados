```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# matplot > pyplot > seaborn
```

```python
# parsed_dates faz com que o Pandas "entenda" que cada linha se refere a uma data, não a um número de linha
df = pd.read_csv("WorldCups.csv", index_col=0, parse_dates=True)
```

```python
df.describe()
```

```python
# dimensão da figura
plt.figure(figsize=(20, 6))

# título
plt.title("Números das Copas do Mundo")

# índice como eixo x, gols marcados como eixo y
sns.barplot(x=df.index, y=df["GoalsScored"])

# esse comando desenha tudo que foi feito até esse ponto
# plt.show()

# nome do eixo y
plt.ylabel("Total de Gols")
```

```python
df = df.loc[:, ["GoalsScored", "MatchesPlayed", "QualifiedTeams"]]

plt.figure(figsize=(14, 7))

plt.title("Números das Copas")

# 'annot' garante que os valores das células apareçam no mapa de calor
sns.heatmap(data=df, annot=True)

plt.xlabel("Medidas")
```
